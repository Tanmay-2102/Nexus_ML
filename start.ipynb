##### ID: Unique identifier for each entry.
##### TargetID: Identifier for the target.

##### DRUGNAME: Name of the drug.
##### PUBCHCID: PubChem compound ID.
##### Disease_of_highest_status: Disease associated with the highest status.
##### Drug_Status: Status of the drug in the dataset.
##### UNIPROID: UniProt identifier for the target protein.
##### TARGNAME: Name of the target.
##### GENENAME: Gene name associated with the target.
##### SYNONYMS: Alternate names for the target.
##### FUNCTION: Biological function of the target.
##### BIOCLASS: Biological classification of the target.
##### SEQUENCE: Amino acid sequence of the target protein.
##### Disease: Disease associated with the drug.
##### Accession Number: Accession number in relevant databases.
##### Target_Status: Current status of the target (e.g., Approved).
